## 📌 題目 1：直方圖 (20 分)

從正態分布產生 1000 筆資料，畫 **bins=30** 的直方圖。

其他要求: 

> title: Normal Distribution Histogram

> xlabel: Value

> ylabel: Frequency

---

````python
# 你的程式

import numpy as np
import matplotlib.pyplot as plt

data = np.random.normal(0, 1, 1000)
plt.hist(data, bins=30, color='skyblue', edgecolor='black')

plt.title('Normal Distribution Histogram')
plt.xlabel('Value')
plt.ylabel('Frequency')


plt.show()
````

````python
# 你的程式

import numpy as np
import matplotlib.pyplot as plt

# 固定亂數種子以便結果可重現
np.random.seed(768)
n = 200

# 散佈圖資料
x = np.random.normal(loc=2.0, scale=1.0, size=n)
y = x * 0.5 + np.random.normal(scale=0.8, size=n)
size_var = np.abs(np.random.normal(loc=20, scale=10, size=n))  # 基礎點面積

# 為了視覺效果，對 size_var 做適度縮放（matplotlib 的 s 參數是面積）
size_scale = size_var * 3 


scatter = plt.scatter(x, y, c=y, s=size_scale, cmap='viridis', alpha=0.7, edgecolor='black')


cbar = plt.colorbar(scatter)
cbar.set_label('y-values')


plt.title('Scatter with Colormap')
plt.xlabel('x')
plt.ylabel('y')

# 顯示圖形
plt.show()
````
